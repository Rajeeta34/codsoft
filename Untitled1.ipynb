```python
import kagglehub

path = kagglehub.dataset_download("adrianmcmahon/imdb-india-movies")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os
df = pd.read_csv(os.path.join(path, "IMDb Movies India.csv"), encoding='latin1')
```

```python
print("\n================ DATASET INFORMATION ================\n")
print(df.info())

print("\n================ FIRST 5 ROWS ================\n")
print(df.head())

print("\n================ SHAPE ================\n")
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
```

```python
print("\n=================== MISSING VALUES ===================\n")

missing = df.isnull().sum()
missing_percent = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing,
    'Percentage': missing_percent
})

print(missing_df)
```

```python
print("\n=================== UNIQUE VALUES ===================\n")

for col in df.columns:
  print(f"{col} : {df[col].nunique()} unique values")
```

```python
if 'Year' in df.columns:
  df['Year'] = df['Year'].astype(str)
  df['Year'] = df['Year'].str.extract(r'(\d{4})')
  df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
```

```python
if 'Duration' in df.columns:
  df['Duration'] = df['Duration'].astype(str)
  df['Duration'] = df['Duration'].str.replace('min', '', regex=False)
  df['Duration'] = pd.to_numeric(df['Duration'], errors='coerce')
```

```python
if 'Votes' in df.columns:
  df['Votes'] = df['Votes'].astype(str)
  df['Votes'] = df['Votes'].str.replace(',', '')
  df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
```

```python
print("\n================= NUMERICAL SUMMARY ================\n")

print(df.describe())
```

```python
if 'Genre' in df.columns:

  print("\n================ TOP GENRES ================\n")

  genre_counts = df['Genre'].value_counts().head(10)
  print(genre_counts)

  plt.figure(figsize=(10,5))
  genre_counts.plot(kind='bar')

  plt.title("Top 10 Genres")
  plt.xlabel("Genre")
  plt.ylabel("Count")

  plt.xticks(rotation=45)
  plt.tight_layout()
  plt.show()
```

```python
if 'Rating' in df.columns:

  plt.figure(figsize=(8,5))

  sns.histplot(df['Rating'], bins=10, kde=True)

  plt.title("Rating Distribution")
  plt.xlabel("Rating")
  plt.ylabel("Count")

  plt.show()
```

```python
if 'Votes' in df.columns:

  plt.figure(figsize=(8,5))

  sns.histplot(df['Votes'], bins=20)

  plt.title("Votes Distribution")
  plt.xlabel("Votes")
  plt.ylabel("Count")

  plt.show()
```

```python
if 'Duration' in df.columns:

  plt.figure(figsize=(8,5))

  sns.histplot(df['Duration'], bins=20, kde=True)

  plt.title("Movie Duration Distribution")
  plt.xlabel("Duration (Minutes)")
  plt.ylabel("Count")

  plt.show()
```

```python
if 'Year' in df.columns:

  plt.figure(figsize=(10,5))

  sns.histplot(df['Year'], bins=30)

  plt.title("Movie Release Per Year")
  plt.xlabel("Year")
  plt.ylabel("Count")

  plt.show()
```

```python
if 'Director' in df.columns:

  print("\n================= TOP DIRECTORS ================\n")

  top_directors = df['Director'].value_counts().head(10)

  print(top_directors)

  plt.figure(figsize=(10,5))

  top_directors.plot(kind='bar')

  plt.title("Top 10 Directors")
  plt.xlabel("Director")
  plt.ylabel("Movies Count")

  plt.xticks(rotation=45)
  plt.tight_layout()

  plt.show()

```

```python
if 'Actor 1' in df.columns:

  print("\n================ TOP ACTORS ================\n")

  top_actors = df['Actor 1'].value_counts().head(10)

  print(top_actors)

  plt.figure(figsize=(10,5))

  top_actors.plot(kind='bar')

  plt.title("Top 10 Lead Actors")
  plt.xlabel("Actor")
  plt.ylabel("Movies Count")

  plt.xticks(rotation=45)
  plt.tight_layout()

  plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)

if not numeric_df.empty:

  plt.figure(figsize=(8,5))

  sns.heatmap(
      numeric_df.corr(),
      annot=True,
      cmap='coolwarm'
  )

  plt.title("Correlation Heatmap")

  plt.show()
```

```python
if 'Rating' in df.columns:

  print("\n================ TOP RATED MOVIES ================\n")

  top_movies = df.sort_values(by='Rating', ascending=False)

  print(top_movies[['Name', 'Rating']].head(10))
```

```python
if 'Votes' in df.columns:

  print("\n================ MOST VOTED MOVIES ================\n")

  voted_movies = df.sort_values(by='Votes', ascending=False)

  print(voted_movies[['Name', 'Votes']].head(10))
```

```python
df.to_csv("Cleaned_IMDb_Movies_India.csv",index=False)

print("\nCleaned dataset saved successfully!")
```
