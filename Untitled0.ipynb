```python
import kagglehub

path = kagglehub.dataset_download("yasserh/titanic-dataset")
print("Path to dataset files:",path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(f"{path}/Titanic-Dataset.csv")
```

```python
print("Shape of Dataset:")
print(df.shape)

print("\n Column Names:")
print(df.columns)

print("\n Dataset Info:")
print(df.info())
```

```python
print("\n Summary Statistics:")
print(df.describe())

print("\n Missing Values:")
print(df.isnull().sum())
```

```python
print("\n Passenger id Summary")
print("------------------------")

print("valid:",df['PassengerId'].count())
print("Missing:",df['PassengerId'].isnull().sum())

print("Mean:",df['PassengerId'].mean())
print("Std Deviation:",df['PassengerId'].std())

print("\n Quantiles:")
print(df['PassengerId'].quantile([0,0.25,0.5,0.75,1]))
```

```python
print("\n Survived Summary")
print("-----------------------")

print(df['Survived'].value_counts())
```

```python
print("\n Passenger Class Summary")
print("-------------------------")

print(df['Pclass'].value_counts())
```

```python
print("\n Gender Summary")
print("-----------------")

print(df['Sex'].value_counts())
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['PassengerId'], bins=10)
plt.title("PassengerId Distribution")
plt.xlabel("PassengerId")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Survived'], bins=2)
plt.title("Survived Distribution")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Pclass'], bins=3)
plt.title("Passenger Class Distribution")
plt.xlabel("Pclass")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Age'].dropna(), bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Fare'], bins=10)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
print("\n Age Group Counts")
print("-----------------------")

age_bins = pd.cut(df['Age'], bins=10)
print(age_bins.value_counts().sort_index())

print("\n Fare Group Counts")
print("------------------------")

fare_bins = pd.cut(df['Fare'], bins=10)
print(fare_bins.value_counts().sort_index())
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Sex', hue='Survived', data=df)

plt.title("Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Pclass', hue='Survived', data=df)

plt.title("Survival by Passenger Class")
plt.show()
```
